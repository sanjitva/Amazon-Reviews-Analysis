```python
import numpy as np
import pandas as pd
```

## Importing Data

```python
df = pd.read_csv('Shoes_Data.csv')
```

## Making a dataframe with every individual review

```python
all_reviews = []
product_title = []
product_id = []
all_ratings = []
all_nreviews = []

for a in list(range(len(df))):
    
    row = df.iloc[a]
    
    review_list = [a for a in row['reviews'].split('|| ')]
    for b in review_list:
        all_reviews.append(b)
        product_title.append(row['title'])
        product_id.append(a)
        all_ratings.append(float(row['rating'].split(' out of 5')[0]))
        all_nreviews.append(int(row['total_reviews'].split(' rating')[0]))
        

        
#Initiating a Pandas DataFrame out of the lists generated above
review_df = pd.DataFrame({'Product ID':product_id,
                  'Product Name':product_title,
                  'No. of Reviews': all_nreviews,
                  'Review': all_reviews,
                  'Product Avg Rating': all_ratings
                 })       
```
